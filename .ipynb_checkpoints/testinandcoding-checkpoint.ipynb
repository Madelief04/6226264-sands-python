```python

import numpy as np

# sine wave with parameters frequency, amplitude, duration, sampling_rate
def generate_sine_wave (frequency, amplitude, duration, sampling_rate): 
    t = np.linspace(0, duration, int(sampling_rate * duration))
    signal = amplitude * np.sin(2 * np.pi * frequency * t)
    return signal

sine_wave = generate_sine_wave (2, 1, 2, 100)

# sawtooth wave with parameters frequency, amplitude, duration, sampling_rate, width =1
def generate_sawtooth_wave (frequency, amplitude, duration, sampling_rate, width =1 ): 
    t = np.linspace(0, duration, int(sampling_rate * duration))
    sawtooth_wave = amplitude * signal.sawtooth(2 * np.pi * frequency * t, width)
    return sawtooth_wave

sawtooth_wave = generate_sawtooth_wave (3, 0.4, 2, 100)

#function for time shift
def time_shift_signal(signal, shift_seconds, sampling_rate):
    shift_samples = int(shift_seconds * sampling_rate)
    return (signal, shift_samples)

sine_shifted = time_shift_signal(sine_wave, 1, 5) 
sawtooth_shifted = time_shift_signal(sawtooth_wave, 3, 100)  

#function for time scaling
def time_scale_signal(signal, scale_factor, sampling_rate):
    original_length = len(signal)
    new_length = int(original_length / scale_factor)
    
    original_indices = np.arange(original_length)
    new_indices = np.linspace(0, original_length - 0.05, new_length)
    
    scaled_signal = np.interp(new_indices, original_indices, signal)
    return scaled_signal

sine_fast = time_scale_signal(sine_wave, 2, 100)
sine_slow = time_scale_signal(sine_wave, 0.5, 100)

sawtooth_fast = time_scale_signal(sawtooth_wave, 2, 100)
sawtooth_slow = time_scale_signal(sawtooth_wave, 0.5, 100)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

#time arrays for plotting
t_sine = np.linspace(0, 2, len(sine_wave))
t_sawtooth = np.linspace(0, 2, len(sawtooth_wave))

#time arrays for scaled signals
t_sine_fast = np.linspace(0, 1, len(sine_fast))  
t_sine_slow = np.linspace(0, 4, len(sine_slow))  
t_sawtooth_fast = np.linspace(0, 1, len(sawtooth_fast))
t_sawtooth_slow = np.linspace(0, 4, len(sawtooth_slow))

plt.figure(figsize=(15, 12))

#original signals
plt.subplot(3, 2, 1)
plt.plot(t_sine, sine_wave)
plt.title('Original Sine Wave (2Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 2, 2)
plt.plot(t_sawtooth, sawtooth_wave)
plt.title('Original Sawtooth Wave (3Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

#time scaled sine waves
plt.subplot(3, 2, 3)
plt.plot(t_sine, sine_wave, 'b-', alpha=0.7, label='Original')
plt.plot(t_sine_fast, sine_fast, 'r-', label='Fast (2x)')
plt.plot(t_sine_slow, sine_slow, 'g-', label='Slow (0.5x)')
plt.title('Sine Wave Time Scaling')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Time scaled sawtooth waves
plt.subplot(3, 2, 4)
plt.plot(t_sawtooth, sawtooth_wave, 'b-', alpha=0.7, label='Original')
plt.plot(t_sawtooth_fast, sawtooth_fast, 'r-', label='Fast (2x)')
plt.plot(t_sawtooth_slow, sawtooth_slow, 'g-', label='Slow (0.5x)')
plt.title('Sawtooth Wave Time Scaling')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Time shifted signals (note: your time_shift_signal returns tuple, so we extract the signal)
plt.subplot(3, 2, 5)
sine_signal, sine_shift = sine_shifted
plt.plot(t_sine, sine_wave, 'b-', alpha=0.7, label='Original')

# For shifted signal, we need to handle the circular shift manually
shifted_sine = np.roll(sine_signal, sine_shift)
plt.plot(t_sine, shifted_sine, 'r-', label=f'Shifted ({sine_shift/sine_signal.size*2:.1f}s)')
plt.title('Sine Wave Time Shift')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.subplot(3, 2, 6)
sawtooth_signal, sawtooth_shift = sawtooth_shifted
plt.plot(t_sawtooth, sawtooth_wave, 'b-', alpha=0.7, label='Original')
shifted_sawtooth = np.roll(sawtooth_signal, sawtooth_shift)
plt.plot(t_sawtooth, shifted_sawtooth, 'r-', label=f'Shifted ({sawtooth_shift/sawtooth_signal.size*2:.1f}s)')
plt.title('Sawtooth Wave Time Shift')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('signal_processing_results.png', dpi=300, bbox_inches='tight')
plt.show()
```
